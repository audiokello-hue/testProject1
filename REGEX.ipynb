```python
### pWorking with REGEX

### search for patterns

import re

example1 = "it rained cats and dogs and my cat was very scared of caterpillar"

pattern = "dogs"

match = re.search(pattern, example1)

if match:
  print("pattern successfully found")
else:
  print("pattern is missing")

```

```python
## extracting information

example1 = "it rained cats and dogs and my cats were very scared of caterpillar"

pattern = 'cat'

re.findall(pattern, example1)
```

```python
### Replacing text

text = "Hello World"

pattern = r"World"
re.sub(pattern, "Mary", text)
```

```python
### Splitting a string

text = "apple,banana,cherry,pineaple"

pattern = r","

re.split(pattern,text)
```
